```python
c1 = []
c2 = []

for line in raw_data:
    valX, valC = line.split(" ")
    valX = float(valX)
    valC = int(float(valC))
    
    T = c1 if valC != 1 else c2
    T.append(valX)
```

```python
# helper functions
import math

def mean(array):
    return sum(array) / len(array)

def standard_deviation(array):
    m = mean(array)
    return sum((x - m) ** 2 for x in array) / len(array)
```

```python
n_c1, mean_c1, stdev_c1 = len(c1), mean(c1), standard_deviation(c1)
n_c2, mean_c2, stdev_c2 = len(c2), mean(c2), standard_deviation(c2)
```
