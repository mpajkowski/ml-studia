```python
raw_data = open("dane.txt").read().strip().split("\n")
```

```python
c1 = []
c2 = []

for line in raw_data:
    valX, valC = line.split()
    valX = float(valX)
    valC = int(float(valC))
    
    T = c2 if valC != 1 else c1
    T.append(valX)
```

```python
%matplotlib notebook

import matplotlib.pylab as plt
plt.hist((c1, c2), 100)
```

```python
# helper functions
import math

def mean(array):
    return sum(array) / len(array)
  
def standard_deviation(array):
    m = round(mean(array), 3)
    return round(math.sqrt(sum((i - m) ** 2 for i in array) / (len(array) - 1)), 3)

def computeDF(arr1, arr2):
    s1 = standard_deviation(arr1)
    s2 = standard_deviation(arr2)
    
    s1_squared = s1 ** 2
    s2_squared = s2 ** 2
    s1_squared = round(s1_squared, 3)
    s2_squared = round(s2_squared, 3)
    n1 = len(arr1)
    n2 = len(arr2)

    DF = (s1_squared / n1 + s2_squared / n2) ** 2 \
       / (( (s1_squared / n1) ** 2 / (n1 - 1)) \
       + ( (s2_squared / n2)**2 / (n2 - 1)))

    return round(DF, 3)
```

```python
mean_c1, stdev_c1 = mean(c1), standard_deviation(c1)
mean_c2, stdev_c2 = mean(c2), standard_deviation(c2)

z = round((mean_c1 - mean_c2) / math.sqrt(stdev_c1 / len(c1) + stdev_c2 / len(c2)), 3)
df = computeDF(c1, c2)

mean_c1 = round(mean_c1, 3)
mean_c2 = round(mean_c2, 3)
stdev_c1 = round(stdev_c1, 3)
stdev_c2 = round(stdev_c2, 3)
```

```python
print("1st sample size = " + str(len(c1)))
print("2nd sample size = " + str(len(c2)))

print("1st sample mean = " + str(mean_c1))
print("2nd sample mean = " + str(mean_c2))
print("1st sample stdev = " + str(stdev_c1))
print("2nd sample stdev = " + str(stdev_c2))
print("Degrees of freedom for Student-t distribution = " + str(df))

print("\nResults:")

print("Statistics value z = " + str(z))
print("Normal distribution: p-value = " + str((1 - 0.99831) * 2)) # z tabeli
print("Student-t distribution: p-value = ")
print("Normal distribution: ")
print("Student-t distribution: ")
```
